# NSynth Pre-Processing

- Download NSynth Data from: https://magenta.tensorflow.org/datasets/nsynth#files
- Split into train and test sets
- Calculate spectrogram and wavelet features

```python
## Download dataset
import urllib.request
urllib.request.urlretrieve('http://download.magenta.tensorflow.org/datasets/nsynth/nsynth-train.jsonwav.tar.gz',
                           'nsynth.tar.gz')
## Extract files from tar
import tarfile
tar = tarfile.open('nsynth.tar.gz')
tar.extractall()
tar.close()
```

```python
## List all files in directory and subdirectories and extract .wav files
import os
from os.path import isfile, join

file_list = []
for path, subdirs, files in os.walk("nsynth-train/audio/"):
    for name in files:
        root, ext = os.path.splitext(name)
        if ext == '.wav':
#             print(os.path.join(path, name))
            file_list.append(os.path.join(path, name))
```

## Calculate Spectrogram Features

- Only work with 50,000 accoustic instruments (most similar to real orchestral audio)
- Calculate following spectrogram features (from raw audio with same hop length as mel-spectrograms):
    - RMS
    - Spectral Centroid
    - Spectral Bandwidth
    - Spectral Flatness
    - Spectral Rolloff
    - Mean Power over Time
    - "Time to Black": time at which each frequency bin hits a dB of -80, -75, -70, etc.
    
Drop ``Bass`` and ``Organ`` instrument families, since not very many examples
Then, 8 Instrument families:
1. Brass (1)
2. Flute (2)
3. Guitar (3)
4. Keyboard (4)
5. Mallet (5)
6. Reed (7)
7. String (8)
8. Vocal (10)

```python
## Select accoustic instruments only
ainds = np.where(source == 0)[0]
print(len(ainds))
## Drop organ and bass
ainds = ainds[np.where((instrument_family[ainds] != 0 ) & 
                       (instrument_family[ainds] != 6))]
ainds = np.random.choice(ainds, 50000, replace = False)
print(len(ainds))
```

```python
instrument_family = instrument_family[ainds]
instrument = instrument[ainds]
pitch = pitch[ainds]
velocity = velocity[ainds]
sound_qualities = sound_qualities[ainds,:]
source = source[ainds]
```

```python
print(np.unique(instrument_family))
np.round(np.unique(instrument_family, return_counts = True)[1]/len(instrument_family),3)
```

```python
rms_list = np.array([librosa.feature.rms(y=audio_list[i], hop_length=502) 
                     for i in ainds])
cs_list = np.array([librosa.feature.spectral_centroid(y=audio_list[i], hop_length=502) 
                    for i in ainds])
bw_list = np.array([librosa.feature.spectral_bandwidth(y=audio_list[i], hop_length=502) 
                    for i in ainds])
fl_list = np.array([librosa.feature.spectral_flatness(y=audio_list[i], hop_length=502) 
                    for i in ainds])
roll_list = np.array([librosa.feature.spectral_rolloff(y=audio_list[i], hop_length=502) 
                      for i in ainds])
```

```python
roll_list.shape
```

```python
rms_list = rms_list[:,0,:]
cs_list = cs_list[:,0,:]
bw_list = bw_list[:,0,:]
fl_list = fl_list[:,0,:]
roll_list = roll_list[:,0,:]
```

```python
stft_list = np.array([librosa.feature.melspectrogram(y=audio_list[i], sr=16000, 
                                                     n_mels=128, fmax=8000, 
                                                     hop_length=502) 
                          for i in ainds])
```

```python
# to dB
stft_list_dB = np.array([librosa.power_to_db(stft_list[i,:,:], ref=np.max) 
                          for i in range(stft_list.shape[0])])
```

```python
stft_list.shape
```

```python
def get0_power(a, threshold):
    mint = -80 ## minimum threshold
    if threshold == mint:
        idx = np.where(a == threshold)[0]
    else:
        idx = np.where(a <= threshold)[0]
    if len(idx) > 1:
        idx = idx[0]
    elif len(idx) == 0:
        idx = 128#len(a)
    return(idx)
```

```python
## Calculate additional features
## median power vs. time
medp_time = np.array([np.apply_along_axis(np.median,  0, stft_list_dB[i,:,:]) 
                      for i in range(stft_list.shape[0])])
## mean power vs. time
meanp_time = np.array([np.apply_along_axis(np.mean,  0, stft_list_dB[i,:,:]) 
                       for i in range(stft_list.shape[0])])
## time to power of -80
tp80 = np.array([np.apply_along_axis(get0_power,  1, stft_list_dB[i,:,:], -80).flatten() 
                 for i in range(stft_list.shape[0])])
tp75 = np.array([np.apply_along_axis(get0_power,  1, stft_list_dB[i,:,:], -75).flatten() 
                 for i in range(stft_list.shape[0])])
tp70 = np.array([np.apply_along_axis(get0_power,  1, stft_list_dB[i,:,:], -70).flatten() 
                 for i in range(stft_list.shape[0])])
```

## Calculate Wavelet Features
- mean wavelet coefficients over frequencies

```python
from scipy import signal
## Look at widths of 1, 5, 10, 25
meanvalF = np.zeros((4, stft_list_dB.shape[0], stft_list_dB.shape[1]))


for j in range(stft_list.shape[0]):
    meanvalF[0, j,:] = [np.mean(signal.cwt(stft_list_dB[j, i, :], signal.ricker,
                    [1])[0]) for i in range(stft_list_dB.shape[1])]
    meanvalF[1, j,:] = [np.mean(signal.cwt(stft_list_dB[j, i, :], signal.ricker,
                    [5])[0]) for i in range(stft_list_dB.shape[1])]
    meanvalF[2, j,:] = [np.mean(signal.cwt(stft_list_dB[j, i, :], signal.ricker,
                    [10])[0]) for i in range(stft_list_dB.shape[1])]
    meanvalF[3, j,:] = [np.mean(signal.cwt(stft_list_dB[j, i, :], signal.ricker,
                    [25])[0]) for i in range(stft_list_dB.shape[1])]
       
```

## Additional Wavelet Features

```python
from scipy import signal
from scipy.stats import kurtosis
## Look at widths of 25
addwavelets = np.zeros((6, stft_dB.shape[0], stft_dB.shape[1]))

for j in range(stft_dB.shape[0]):
    wavetransform = np.array([signal.cwt(stft_dB[j, i, :], signal.ricker,
                    [25])[0] for i in range(stft_dB.shape[1])])
    
    addwavelets[0, j,:] = np.median(wavetransform, axis = 1)
    addwavelets[1, j,:] = np.std(wavetransform, axis = 1)
    addwavelets[2, j,:] = np.var(wavetransform, axis = 1)
    addwavelets[3, j,:] = kurtosis(wavetransform, axis = 1)
    addwavelets[4, j,:] = np.quantile(wavetransform, 0.25, axis = 1)
    addwavelets[5, j,:] = np.quantile(wavetransform, 0.75, axis = 1)
    if j % 1000 == 0:
        print(j)
   
```
